2. How many total unique words are in the corpus?  (Please feel free to define unique words in any interesting, defensible way).

```python
nltk.word_tokenize("hello world")
```

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
emma = nltk.corpus.gutenberg.words('austen-emma.txt')
emma = [w for w in emma if w.isalpha()]
from nltk.corpus import stopwords
emma = [w for w in emma if w not in stopwords.words('english')]
```

```python
emma_unique = set(emma)
len(emma_unique)
```

3. Taking the most common words, how many unique words represent half of the total words in the corpus?Identify the 200 highest frequency words in this corpus.

```python
freq_emma = nltk.FreqDist(emma)
emma_200 = pd.DataFrame(freq_emma.most_common(200),columns=['word','count'])
emma_200['rel_freq'] = emma_200['count']/float(len(emma))
emma_200.head(200)
```

```python
emma_200 = pd.DataFrame(freq_emma.most_common(200),columns=['word','count'])
emma_200['rel_freq'] = emma_200['count']/float(len(emma))
```

```python
len(emma)/2.0 
```

```python
sum(emma_200[:200]['rel_freq'])
```

4. Create a graph that shows the relative frequency of these 200 words.

```python
freq_emma.plot(200, cumulative=True,color='orangered')
```

5. Does the observed relative frequency of these words follow Zipf’s law? Explain.
6. In what ways do you think the frequency of the words in this corpus differ from “all words in all corpora.

```python
g = sns.barplot(x=emma_200.word, y=emma_200.rel_freq)
```

The Zipf's Law in the frequency of any word is approximately inversely proportional to its ranking in the frequency table. 
This law holds for all words in all corpora. 
The only differences are the words themselves,they differ based on the content of the corpus being analyzed.
